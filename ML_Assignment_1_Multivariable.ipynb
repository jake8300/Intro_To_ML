```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#
df = pd.read_csv('./D3.csv')
df.head(100)
dataset = df.values[:,:]
print(dataset[:20,:])
X = df.values[:,0:3]
Y = df.values[:,3]
```

```python
#Convert 1D to 2D array
m = len(X)
X_1 = X.reshape(m,3)
print("X_1 =", X_1[:5,:])


m = len(X)
X_0 = np.ones((m,1))
X_0[:5], len(X_0)

X = np.hstack((X_0, X_1))
X[:5]

theta = np.zeros((4,1))
theta
```

```python
def compute_loss(X, Y, theta):

    Y_Pred = X.dot(theta)
    Error= np.subtract(Y_Pred, Y)
    Sqr_error = np.square(Error)
    J = 1 / (2 * m) * np.sum(Sqr_error)

    return J
```

```python

def gradient_descent(X, Y, theta, alpha, iterations):

    Previous_loss = np.zeros(iterations)

    for i in range(iterations):

        Y_Pred = X.dot(theta)
        Error = np.subtract(Y_Pred, Y)
        sum_delta = (alpha / m) * X.transpose().dot(Error);
        theta = theta - sum_delta;
        Previous_loss[i] = compute_loss(X, Y, theta)

    return theta, Previous_loss

theta = [0., 0., 0., 0.]
iterations = 100
alpha = 0.01
```

```python
theta, loss_history = gradient_descent(X, Y, theta, alpha, iterations)
plt.plot(range(1, iterations + 1), loss_history)

plt.title("Loss over the iteration")
plt.ylabel("Loss")
plt.xlabel("Iterations")
```

```python
theta_0 = theta[0]
theta_1 = theta[1]
theta_2 = theta[2]
theta_3 = theta[3]
```

```python
# Y Prediciton for (1, 1, 1)

new_X1 = 1
new_X2 = 1
new_X3 = 1

Y_pred_1 = new_X3*theta_3 + new_X2*theta_2 + new_X1*theta_1 + theta_0*1
print("Predicted Y value for (1,1,1) = ",Y_pred_1)
```
