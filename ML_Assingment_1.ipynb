```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('D3.csv')
df.head(100)
data=df.values[:,:]
plt.rcParams["figure.figsize"] = [10,10]
Y = df.values[:,3]
```

Setup Code above

Below is the start of the X1 code

```python
X1 = df.values[:,0]

plt.scatter(X1,Y,)
plt.title('Training Data for X1')
plt.ylabel('Y value')
plt.xlabel('X1 value')
```

```python
#convert 1D to 2D array
m_1 = len(X1)
X_1 = X1.reshape(m_1,1)

X_0 = np.ones((m_1,1))

X1 = np.hstack((X_0, X_1))
theta_1 = np.zeros((2,1))
```

```python
def compute_loss_1(X1, Y, theta_1):
    Y_Pred_1 = X1.dot(theta_1)
    error_1= np.subtract(Y_Pred_1, Y)
    Sqr_error_1 = np.square(error_1)
    J_1 = 1 / (2 * m_1) * np.sum(Sqr_error_1)

    return J_1
```

```python
def Gradient_D_1(X_1, Y, theta_1, Learn_Rate_1, iterations_1):
    Previous_loss_1 = np.zeros(iterations_1)

    for i_1 in range(iterations_1):

        Y_Pred_1 = X1.dot(theta_1)

        error_1 = np.subtract(Y_Pred_1, Y)
        sum_delta_1 = (Learn_Rate_1 / m_1) * X1.transpose().dot(error_1);
        theta_1 = theta_1 - sum_delta_1;
        Previous_loss_1[i_1] = compute_loss_1(X1, Y, theta_1)

    return theta_1, Previous_loss_1

theta_1 = [0., 0.]
iterations_1 = 100
Learn_Rate_1 = 0.01
theta_1, Previous_loss_1 = Gradient_D_1(X1, Y, theta_1, Learn_Rate_1, iterations_1)
```

```python
theta_1 = [0., 0.]
iterations_1 = 100
Learn_Rate_1 = 0.01
theta_1, Previous_loss_1 = Gradient_D_1(X1, Y, theta_1, Learn_Rate_1, iterations_1)

#Testing
print(X1[:,1][0])
print()
print(X1.dot(theta_1)[0])
print()
print(X1[:,1].min(), X1.dot(theta_1).min())
print()
print(X1[:,1].max(), X1.dot(theta_1).max())
print()
print('m = ')
print(((X1.dot(theta_1).min())-(X1.dot(theta_1).max())) / ((X1[:,1].max())-(X1[:,1].min())))
print()
print('c = ')
print((X1.dot(theta_1).max())-(X1[:,1].min()*-0.21891654486322576))
```

```python
# Graphing linear regression with training data points

plt.scatter(X1[:,1], Y)
plt.plot(X1[:,1], X1.dot(theta_1), color='Red')
plt.title("Final Regression Model for X1")
plt.ylabel("Y")
plt.xlabel("X1")
```

```python
#Loss over iterations for X1

plt.plot(range(1, iterations_1 + 1), Previous_loss_1)
plt.title("Loss over Iterations for X1")
plt.ylabel("Loss")
plt.xlabel("Iterations")
```
